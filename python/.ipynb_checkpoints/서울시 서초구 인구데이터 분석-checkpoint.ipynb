```python
import pandas as pd
import matplotlib.pyplot as plt

result = pd.read_csv('201312.csv', encoding='cp949')

data={}

for i in range(1, len(result)):
    data[result.iloc[i,0]] = result.iloc[i, range(1, len(result.columns), 6)]
    
# 시각화 처리
plt.ylim([400000,500000])

loc = '서울특별시 서초구 (1165000000)'

for i in range(len(data[loc])):
    data[loc][i] = int(data[loc][i].replace(',',''))
    plt.bar(data[loc].index[i].split('년_')[0], data[loc][i])
    
plt.title('2013~2022 서초구 인구변화 추이')
plt.show()
```

```python
# 서초구 남자수
print(result.iloc[22, range(4, len(result.columns), 6)])

# 서초구 여자수
print(result.iloc[22, range(5, len(result.columns), 6)])

# 서초구 남여비율
print(result.iloc[22, range(6, len(result.columns), 6)])
```

```python
figure = plt.figure()
axes1 = figure.add_subplot(111)

x1 = result.columns[range(5, len(result.columns), 6)]
y1= result.iloc[22, range(5, len(result.columns), 6)]
axes1.bar(x1,y1, color='red')
plt.xticks(rotation=90)

plt.title('2013~2022 서초구 여자 인구변화추이')
plt.show()
```

```python
figure = plt.figure()
axes2 = figure.add_subplot(111)

x2 = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
y2= result.iloc[22, range(4, len(result.columns), 6)]
axes2.bar(x2,y2, color='b')
plt.xticks(rotation=90)

plt.title('2013~2022 서초구 남자 인구변화추이')
plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

label = result.columns[range(6, len(result.columns), 6)]
data = result.iloc[22, range(6, len(result.columns), 6)]

axes.pie(data, labels = label)

plt.show()
```
