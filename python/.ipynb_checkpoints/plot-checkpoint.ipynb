```python
x = [1,2,3,4]
y = [2,4,6,8]

plt.plot(x,y)
plt.show()
```

```python
import numpy as np
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x = [1,2,3,4]
y = [2,4,6,8]

x2 = np.array([1,2,3,4])
y2 = np.array([4,1,3,1])

axes.plot(x,y, color='red', linestyle='dotted', marker = 'o')
axes.plot(x2,y2, color='green', linestyle='dashed', marker='^')
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x = [1,3,5,7]
y = [2,4,6,8]

x2 = [2,4,6,8]
y2 = [4,4,3,6]

axes.bar(x,y, color='gray')
axes.bar(x2,y2, color='green')
plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)
axes2 = axes.twinx()

x = [1,2,3,4]
y = [2,4,6,8]

x2 = [1,2,3,4]
y2 = [4,6,1,9]

axes.bar(x,y, color='gray', label='bar')
axes2.plot(x2,y2, color='r', label = 'plot')

axes.legend()
axes2.legend(loc=4)

plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x = [1,2,3,4]
y = [2,4,6,8]

x2 = [1,2,3,4]
y2 = [4,6,1,9]

axes.scatter(x,y, color='gray')
axes.scatter(x2,y2, color='r')

plt.show()
```

```python
from matplotlib import font_manager, rc

font_path = '/Windows/Fonts/batang.ttc'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

figure = plt.figure()
axes = figure.add_subplot(111)

label = ['', '', 'javascript', 'html']
data = [1,2,5,2]

axes.pie(data, labels = label)

plt.show()
# plt.savefig('test')
```
